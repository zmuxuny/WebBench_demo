```python
import matplotlib.pyplot as plt
# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False   # 解决负号 '-' 显示为方块的问题
conc = [50, 100, 200, 400, 800]
rpm = [40947, 88188, 160884, 142944, 118965]  # 每分钟请求数
failed = [0, 2, 20, 10, 31]  # 失败请求数

# 创建图表
fig, ax1 = plt.subplots()

# 左轴：每分钟请求数
ax1.set_xlabel('并发数')
ax1.set_ylabel('每分钟请求数 (pages/min)', color='tab:blue')
ax1.plot(conc, rpm, marker='o', color='tab:blue', label='pages/min')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 右轴：失败请求数
ax2 = ax1.twinx()
ax2.set_ylabel('失败请求数 (Failed Requests)', color='tab:red')
ax2.plot(conc, failed, marker='x', color='tab:red', label='Failed Requests')
ax2.tick_params(axis='y', labelcolor='tab:red')

# 标题和网格
plt.title("并发数 vs 性能指标")
plt.grid(True)

# 显示图例
fig.tight_layout()
plt.show()
```
