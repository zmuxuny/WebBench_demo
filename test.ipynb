```python
import matplotlib.pyplot as plt
# 
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False   #  '-' 
conc = [50, 100, 200, 400, 800]
rpm = [40947, 88188, 160884, 142944, 118965]  # 
failed = [0, 2, 20, 10, 31]  # 

# 
fig, ax1 = plt.subplots()

# 
ax1.set_xlabel('')
ax1.set_ylabel(' (pages/min)', color='tab:blue')
ax1.plot(conc, rpm, marker='o', color='tab:blue', label='pages/min')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 
ax2 = ax1.twinx()
ax2.set_ylabel(' (Failed Requests)', color='tab:red')
ax2.plot(conc, failed, marker='x', color='tab:red', label='Failed Requests')
ax2.tick_params(axis='y', labelcolor='tab:red')

# 
plt.title(" vs ")
plt.grid(True)

# 
fig.tight_layout()
plt.show()
```
